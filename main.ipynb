## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
#your code here
    return 1 if n <= 1 else n * factorial(n-1)

factorial(6)

```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... ðŸ§`

```python
def mode_counter(arr):
#your code here
    arr_sorted = sorted(arr)
    highest_frequency = arr_sorted[0]
    current = arr_sorted[0]
    highest_count = count = 1

    for x in arr_sorted[1:]:
        if x == current:
            count +=1
        else:
            current = x
            count = 1

        if count > highest_count:
            highest_count = count
            highest_frequency = current
            
    return highest_frequency

mode_counter(arr)
```

```python
# This will test your function 
test_mode(mode_counter)
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. ðŸ˜‰`

```python
def st_dev(list_sd):
#your code here
    mean_data = sum(list_sd) / len(list_sd) #calculate the mean
    squared_differences = [(i - mean_data)**2 for i in list_sd] 
    variance = sum(squared_differences) / (len(list_sd)-1)
    sd = variance ** 0.5 #square root
    return sd
```

```python
# This will test your function 
test_stdev(st_dev)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ðŸ¤ª`

```python
def sort_alpha(string):
    words = []
    new = ""
    for i in string:
        if i != ",":
            new += i
        else:
            words.append(new)
            new = ""
    words.append(new)
    words = sorted(words)
    final_list = ""
    for ii in range(len(words)):
        final_list += words[ii]
        if ii != len(words)-1:
            final_list += ","

sort_alpha(string)
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
#your code here
    hasnumber = any(i.isdigit() for i in password) #checks if there is a number in the password
    haslower = any(i.islower() for i in password) #checks if there is a lowercase letter in pw
    hasupper = any(i.isupper() for i in password) #checks if there is an uppercase letter in pw
    islong = len(password) >=8 #checks if len is min 8
    hasspecialchar = any(not i.isalnum() for i in password) #checks if there are any letters that are NOT alphanumeric
    if hasupper == True and haslower == True and hasnumber is True and islong == True and hasspecialchar == True: #returns true if all conditions are met
        return True
    else:
        return False

check_pass(password)
```

```python
# This will test your function 
test_pass(check_pass)
```
